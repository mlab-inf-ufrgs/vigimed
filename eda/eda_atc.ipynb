```python
# Criar dim_atc com estrutura hierárquica de 7 níveis
from utils import build_dim_atc

# Construir dimensão ATC com níveis hierárquicos
dim_atc_hierarquica = build_dim_atc(
    dim_atc,
    atc_code_col='ATC code',
    atc_name_col='ATC level name',
    ddd_col='DDD',
    unit_col='Unit',
    adm_r_col='Adm.R',
    comment_col='Comment'
)

print(f"Shape da dim_atc hierárquica: {dim_atc_hierarquica.shape}")
print(f"\nColunas: {list(dim_atc_hierarquica.columns)}")
print(f"\nPrimeiras linhas:")
dim_atc_hierarquica.head(10)
```
