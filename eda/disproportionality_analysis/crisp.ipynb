```python
%%sql
DROP TABLE IF EXISTS fat_analytics;

CREATE TABLE fat_analytics AS
with fat_med as (
SELECT 
    fat_medicamentos.IDENTIFICACAO_NOTIFICACAO, 
    atc.ATC_CODE_LEVEL_4_LEVEL_NAME,
    (
        CASE 
            WHEN atc.ATC_CODE_LEVEL_4 IN ('L04AB', 'L04AA','L01FA', 'L04AC','L04AF') 
                THEN (
                    CASE 
                        /*Rituximab*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%rituximab%' THEN 'L01FA01_Rituximab'
                        /*Abatacept*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%abatacept%' THEN 'L04AA24_Abatacept'
                        /*Etanercept*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%etanercept%' THEN 'L04AB01_Etanercept'
                        /*Infliximab*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%infliximab%' THEN 'L04AB02_Infliximab'
                        /*Adalimumab*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%adalimumab%' THEN 'L04AB04_Adalimumab'
                        /*Certolizumab Pegol*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%certolizumab%' THEN 'L04AB05_Certolizumab Pegol'   
                        /*Golimumab*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%golimumab%' THEN 'L04AB06_Golimumab'
                        /*Tocilizumab*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%tocilizumab%' THEN 'L04AC07_Tocilizumab'
                        /*Baricitinib */ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%baricitinib%' THEN 'L04AF02_Baricitinib'
                        /*Upadacitinib*/ WHEN LOWER(PRINCIPIOS_ATIVOS_WHODRUG) LIKE '%upadacitinib%' THEN 'L04AF03_Upadacitinib' 
                        ELSE 'Outros'
                    END
                )
            ELSE 'Outros'
        END
    ) AS ATC_CODE_LEVEL_5_LEVEL_NAME
FROM fat_medicamentos
INNER JOIN dim_atc AS atc
    ON fat_medicamentos.ATC_CODE_LEVEL_4 = atc.ATC_CODE_LEVEL_4 
    AND atc.ATC_LEVEL = 4
)
,fat as (
select 
    dim.IDENTIFICACAO_NOTIFICACAO AS ID
    ,dim.SEXO
    ,fat_med.ATC_CODE_LEVEL_4_LEVEL_NAME AS ATC_LEVEL_4
    ,fat_med.ATC_CODE_LEVEL_5_LEVEL_NAME AS ATC_LEVEL_5  
    ,dim_soc_llt.REACAO_EVTO_ADVERSO_MEDDRA_LLT_VALOR AS REACAO_LLT
    ,dim_soc_llt.PT AS REACAO_PT
from dim_notificacoes as dim
inner join fat_med using (IDENTIFICACAO_NOTIFICACAO)
inner join fat_reacoes using (IDENTIFICACAO_NOTIFICACAO)
left join dim_soc_llt on fat_reacoes.LLT_CHAVE = dim_soc_llt.LLT_CHAVE
)

-- Notificações com medicamentos específicos (não 'Outros')
select * from fat 
where ATC_LEVEL_5 <> 'Outros'

union all 

-- Notificações com 'Outros' SOMENTE se não tiverem medicamentos específicos que iremos estudar
select * from fat as a
where a.ATC_LEVEL_5 = 'Outros'
and a.ID NOT IN (
    SELECT b.ID
    FROM fat as b
    WHERE b.ATC_LEVEL_5 <> 'Outros' 
)
```

# **7️⃣ Modelagem**

### Funcoes

```python
import numpy as np
import pandas as pd
from typing import List, Optional


def disproportionality_table(
    df: pd.DataFrame,
    case_col: str = "ID",
    drug_col: str = "ATC_LEVEL_5",
    event_col: str = "REACAO_LLT",
    strata_cols: Optional[List[str]] = None,  # ex.: ["SEXO"]
    min_a: int = 3,
    correction: float = 0.5,
    require_all_methods: bool = False,
) -> pd.DataFrame:
    """
    Tabela 2x2 por (drug, event) em nível de caso (ID) com dedup:
      a = #casos com drug & event
      b = #casos com drug & não-event
      c = #casos com event & não-drug
      d = #casos sem drug & sem event
    Calcula: ROR (LCL95), PRR + chi2, BCPNN (IC/IC025), EBGM (EBGM05).
    """

    if strata_cols is None:
        strata_cols = []

    cols = [case_col, drug_col, event_col] + strata_cols
    data = df.loc[:, cols].dropna(subset=[case_col, drug_col, event_col]).copy()

    # opcional: melhora performance
    for c in [drug_col, event_col] + strata_cols:
        data[c] = data[c].astype("category")

    def _compute_one(sub: pd.DataFrame) -> pd.DataFrame:
        # Dedup caso-droga-evento para não inflar "a" por duplicatas idênticas
        pairs = sub[[case_col, drug_col, event_col]].drop_duplicates()
        if pairs.empty:
            return pd.DataFrame()

        N_total = sub[case_col].nunique()

        case_drug = sub[[case_col, drug_col]].drop_duplicates()
        n_drug = case_drug.groupby(drug_col)[case_col].nunique()

        case_evt = sub[[case_col, event_col]].drop_duplicates()
        n_evt = case_evt.groupby(event_col)[case_col].nunique()

        base = (
            pairs.groupby([drug_col, event_col])[case_col]
            .nunique()
            .rename("a")
            .reset_index()
        )

        base["n_drug"] = base[drug_col].map(n_drug).astype(int)
        base["n_evt"] = base[event_col].map(n_evt).astype(int)

        base["b"] = base["n_drug"] - base["a"]
        base["c"] = base["n_evt"] - base["a"]
        base["d"] = N_total - base["a"] - base["b"] - base["c"]
        base["N_total"] = N_total

        # critério N>=3 (a>=3)
        base = base.loc[base["a"] >= min_a].copy()
        if base.empty:
            return base

        a = base["a"].to_numpy(dtype=float)
        b = base["b"].to_numpy(dtype=float)
        c = base["c"].to_numpy(dtype=float)
        d = base["d"].to_numpy(dtype=float)

        # Correção quando alguma célula for 0
        zero_mask = (a == 0) | (b == 0) | (c == 0) | (d == 0)
        a2 = a + correction * zero_mask
        b2 = b + correction * zero_mask
        c2 = c + correction * zero_mask
        d2 = d + correction * zero_mask

        # ROR
        ror = (a2 * d2) / (b2 * c2)
        se = np.sqrt(1 / a2 + 1 / b2 + 1 / c2 + 1 / d2)
        ror_lcl95 = np.exp(np.log(ror) - 1.96 * se)

        # PRR + chi2
        prr = (a2 / (a2 + b2)) / (c2 / (c2 + d2))
        chi2 = ((a2 * d2 - b2 * c2) ** 2) * (a2 + b2 + c2 + d2) / (
            (a2 + b2) * (c2 + d2) * (a2 + c2) * (b2 + d2)
        )

        # BCPNN (IC, IC025)
        ic = np.log2((a2 * (a2 + b2 + c2 + d2)) / ((a2 + b2) * (a2 + c2)))
        ln2 = np.log(2.0)
        v_ic = (1 / (ln2**2)) * (
            (1 / a2) + (1 / (a2 + b2)) + (1 / (a2 + c2)) - (1 / (a2 + b2 + c2 + d2))
        )
        v_ic = np.clip(v_ic, 0, None)
        ic025 = ic - 2.0 * np.sqrt(v_ic)

        # EBGM (forma do artigo) + limite inferior 95% análogo
        ebgm = (a2 * (a2 + b2 + c2 + d2)) / ((a2 + b2) * (a2 + c2))
        ebgm05 = np.exp(np.log(ebgm) - 1.96 * se)

        base["ROR"] = ror
        base["ROR_LCL95"] = ror_lcl95
        base["PRR"] = prr
        base["CHI2"] = chi2
        base["IC"] = ic
        base["IC025"] = ic025
        base["EBGM"] = ebgm
        base["EBGM05"] = ebgm05

        # critérios
        base["signal_ROR"] = base["ROR_LCL95"] > 1
        base["signal_PRR"] = (base["PRR"] >= 2) & (base["CHI2"] >= 4)
        base["signal_BCPNN"] = base["IC025"] > 0
        base["signal_EBGM"] = base["EBGM05"] > 2

        if require_all_methods:
            base["signal"] = (
                base["signal_ROR"]
                & base["signal_PRR"]
                & base["signal_BCPNN"]
                & base["signal_EBGM"]
            )
        else:
            base["signal"] = (
                base["signal_ROR"]
                | base["signal_PRR"]
                | base["signal_BCPNN"]
                | base["signal_EBGM"]
            )

        return base

    if not strata_cols:
        out = _compute_one(data)
        if out.empty:
            return out
        return out.sort_values(["signal", "ROR_LCL95", "a"], ascending=[False, False, False])

    frames = []
    for keys, sub in data.groupby(strata_cols, dropna=False):
        res = _compute_one(sub)
        if res.empty:
            continue
        if not isinstance(keys, tuple):
            keys = (keys,)
        for col, val in zip(strata_cols, keys):
            res.insert(0, col, val)
        frames.append(res)

    if not frames:
        return pd.DataFrame()

    out = pd.concat(frames, ignore_index=True)
    return out.sort_values(strata_cols + ["signal", "ROR_LCL95", "a"],
                           ascending=[True]*len(strata_cols) + [False, False, False])
 
```
