# 🥈 Silver

normalized data, medium quality


## hist_fat_notificacoes_reacoes

```python
path = Path(project_root) / "data/03_gold/dim_soc/dim_soc.csv"
dim_soc = pd.read_csv(path)  
dim_soc.head()

```

```python
path = Path(project_root) / "data/01_bronze/Reacoes/Reacoes.parquet"
silver = pd.read_parquet(path)  
silver.head()

```

```python
silver.columns
```

```python
# Supondo que no DataFrame `norm` as colunas tenham os mesmos nomes
silver = bronze.merge(
    dim_soc,
    how='left',
    left_on=['SOC', 'HLGT', 'HLT', 'PT', 'REACAO_EVTO_ADVERSO_MEDDRA_LLT'],
    right_on=['SOC', 'HLGT', 'HLT', 'PT', 'REACAO_EVTO_ADVERSO_MEDDRA_LLT']
)


silver["DATA_INICIO_HORA"] = pd.to_datetime(silver["DATA_INICIO_HORA"], format='%Y%m%d', errors='coerce')
silver["DATA_FINAL_HORA"] = pd.to_datetime(silver["DATA_FINAL_HORA"], format='%Y%m%d', errors='coerce')

silver = silver[['IDENTIFICACAO_NOTIFICACAO', 'PK_SOC', 'DATA_INICIO_HORA', 'DATA_FINAL_HORA', 'DURACAO','GRAVE', 'GRAVIDADE', 'DESFECHO']]
 
silver.head()
```

```python
silver.info()
```

```python
silver.to_parquet(Path(project_root) / "data/02_silver/hist_fat_notificacoes_reacoes/hist_fat_notificacoes_reacoes.parquet", index=False)
```
